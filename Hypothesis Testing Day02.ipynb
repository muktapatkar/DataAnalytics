Q1. This dataset has the alcohol consumption before and after love failure. Perform a test to check whether the alcohol consumption is more after the love failure. Alpha =5%

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alcohol=pd.read_csv("C:\\Users\\DELL\\Desktop\\PGPDSE\\Statistics\\Alcohol Consumption.csv")
```

```python
alcohol
```

```python
alcohol.shape
```

```python
plt.scatter(alcohol['Before_BreakUp'],alcohol['After_Breakup'])
plt.show()
```

Step 01: Defining Null and Alternate Hypothesis:
H0: No effect is visualised in before and after drinking habits dure to love failure.
    Avg of Before_Breakup = Avg of After_Breakup
H1: There is an effect visualised after breakup drinking habits. Two tailed test.
    Avg of Before breakup is not equal to Avg of After breakup
    
Step 02: Defining Alpha
    Alpha is 5% which means the confidence level is 95%.
    
Step 03: Defining the test
    1. Population SD is not given which means t-test
    2. Sample size is less n=20.
    3. Sample size is same for both the cases.
   Hence pairwise t-test will be performed.
   

```python
import scipy.stats as stats
import seaborn as sns
```

```python
tstats,pvalue= stats.ttest_rel(alcohol['Before_BreakUp'],alcohol['After_Breakup'])
```

```python
print("The pvalue is", round(pvalue,3))
```

Conclusion: Since the pvalue from the test is greater than the pvalue which is 5%, hence we fail to reject the null hypothesis which means that the alcohol consumption is not influenced by love failure.

P Value: Every t-test comes with P value.A P value is the probabilty that the results from the sampled data occurred by chance. If the P value is higher then it indicates that the data or results have occurred due to random chance, which means that lower P values are good. Lower P values suggest that the data or the results has not occured by random chance. In simple terms, the P value of 0.01 means that there is probabilty of 1% that the results from the experiment happenned due to random chance. The std P value is 5%.

About Pair-wise t-test: Pair-wise t-test appears to be a t-test that helps in performing hypothesis test of two dependent or linked samples coming from the data. Having said that the pair-wise t-test and independent one sample t-test are quite similar in nature.

Q2. The dataset represents the quantity of the soft drink filled in a sample of 50 bottles of 2 liter each. At 5% significance level is there evidence that the mean quantity filled is different from 2 liters.

```python
softdrink
```

```python
softdrink.shape
```

Step 01: Defining Null and Alternate Hypothesis:
H0: The quantity filled in the bottle is 2 liters which means the avg quantity =2
H1: Yes the quantity is either less or more. hence it is two tailed test.

Step 02: Defining alpha:
    Alpha=5% and CF=95%

```python
sns.distplot(softdrink)
```

```python
stat,pvalue=stats.ttest_1samp(softdrink,2)
```

```python
print("P value is", np.round(pvalue,4))
```

Conclusion: Since P value is greater than the critical value which is 5% and hence we fail to reject the null hypothesis which means that the soft drink has quantity of 2 litres.

    H1: The avg body temp of 17 year olds is different than 98.6
Step 02: Defining alpha:
